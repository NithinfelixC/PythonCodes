```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
FIFA_ds = pd.read_csv('C:/Users/hp/OneDrive/Desktop/PythonCodes/Data Set/FIFA19 Player Set Top 20.csv')
FIFA_ds.head()
```

```python
FIFA_ds['Nationality'].value_counts().plot(kind='bar',figsize=(20,5), facecolor='black')
plt.xlabel('Based on Nationality', fontsize=20, color='red')
plt.ylabel('Count of Players', fontsize=20, color='red')
plt.xticks(fontsize=12, rotation=360)
plt.title('FIFA19 Top 20 players Nationality Segmentation', fontsize=30)
```

```python
x = list(FIFA_ds['Nationality'])
y = list(FIFA_ds['Release Clause'])

plt.figure(figsize=(12,5))
plt.bar(x,y,color='green')
plt.title('Top Release Clause value based on Nationality for top 20 players', fontsize=20)
plt.show()
```
