# Plot Chart

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
# While ploting a seperate graph where we are analyzing the Sales vs Fixed Cost. 

x = [1,2,3]
y = [1,20,30]
z = [15,15,15]
plt.plot(x,y)
plt.plot(x,z)
plt.title('Title')
plt.xlabel('X Axis')
plt.ylabel('Y Axis & Z Axis')
plt.legend(["Sales", "Fixed Cost"])
plt.show()
```

```python
# While ploting a seperate graph where we are analyzing the Sales vs Cost. 

x = [0, 5,10,17,25]
y = [0, 5,10,17,25]
z = [13, 13,14,18,20]
r = [13, 13,13,13,13]
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,r)
plt.title('Sales vs Cost', fontdict ={'fontname':'Calibri', 'fontsize':22})
plt.xlabel('X Axis', fontdict ={'fontname':'Calibri', 'fontsize':14})
plt.ylabel('Y Axis', fontdict ={'fontname':'Calibri', 'fontsize':14})
plt.legend(["Sales", "Variable Cost", "Fixed Cost"])

plt.show()
```

```python
# Formating line color, size and background.

a = [0,1,2,3,4,5]
b = [0,3,5,7,9,6]
plt.plot(a,b, label='Sales', color='red', linestyle='dotted', marker='D', markerfacecolor='blue', markeredgecolor='pink')

ax = plt.axes()
ax.set_facecolor("black")
plt.show()
```
